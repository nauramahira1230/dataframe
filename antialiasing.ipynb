```python
mport numpy as np
import matplotlib.pyplot as plt
import math

# Menggenerate segitiga tanpa anti-aliasing
def generate_triangle_normal(image_size, side_length):
    image = np.zeros((image_size, image_size), dtype=np.uint8)
    center_x, center_y = image_size / 2, image_size / 2
    height = (math.sqrt(3) / 2) * side_length  # tinggi segitiga sama sisi

    for i in range(image_size):
        for j in range(image_size):
            # Menggunakan koordinat untuk membentuk segitiga
            dx = j - center_x
            dy = center_y - i
            if abs(dx) <= side_length / 2 and dy <= height and dy >= -math.sqrt(3) * abs(dx) + height:
                image[i][j] = 255

    return image

# Menggenerate segitiga dengan anti-aliasing menggunakan point sampling
def generate_triangle_antialiased(image_size, side_length):
    """Generate triangle with antialiasing using point sampling"""
    image = np.zeros((image_size, image_size), dtype=np.uint8)
    center_x, center_y = image_size / 2, image_size / 2
    height = (math.sqrt(3) / 2) * side_length  # tinggi segitiga sama sisi
    grid_size = 5  # Ukuran grid untuk sampling

    for i in range(image_size):
        for j in range(image_size):
            points_inside = 0
            # Periksa setiap subpixel
            for x in range(grid_size):
                for y in range(grid_size):
                    # Hitung posisi subpixel yang tepat
                    sx = i + (x / grid_size)
                    sy = j + (y / grid_size)
                    dx = sy - center_x
                    dy = center_y - sx
                    # Periksa apakah subpixel berada di dalam segitiga
                    if abs(dx) <= side_length / 2 and dy <= height and dy >= -math.sqrt(3) * abs(dx) + height:
                        points_inside += 1

            # Hitung nilai antialiasing
            antialiasing_value = (points_inside / (grid_size * grid_size)) * 255
            image[i][j] = int(antialiasing_value)

    return image

# Set parameter
image_size = 200
side_length = image_size / 2  # Panjang sisi segitiga

# Generate kedua versi segitiga
normal_triangle = generate_triangle_normal(image_size, side_length)
antialiased_triangle = generate_triangle_antialiased(image_size, side_length)

# Simpan gambar sebagai file JPG
plt.imsave("gkt2_normal_triangle.jpg", normal_triangle, cmap='gray')
plt.imsave("gkt2_antialiased_triangle.jpg", antialiased_triangle, cmap='gray')

# Tampilkan gambar untuk verifikasi
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Tampilkan segitiga normal
ax1.imshow(normal_triangle, cmap='gray', interpolation='nearest')
ax1.set_title('Normal Triangle')
ax1.axis('off')

# Tampilkan segitiga dengan anti-aliasing
ax2.imshow(antialiased_triangle, cmap='gray', interpolation='nearest')
ax2.set_title('Antialiased Triangle')
ax2.axis('off')

plt.show()

berikan penjelasan dalam setiap bagian nya
```
