```python
import numpy as np

# Data pengumpulan sampah (dalam ton)
sampah_wilayah1 = np.array([
    1.2, 1.5, 1.3, 1.7, 1.6, 1.4, 1.1, 1.8, 1.5, 1.9,
    1.3, 1.6, 1.7, 1.4, 1.2, 1.5, 1.3, 1.9, 1.8, 1.7,
    1.5, 1.3, 1.2, 1.4, 1.6, 1.5, 1.9, 1.8, 1.6, 1.7
])

sampah_wilayah2 = np.array([
    1.1, 1.3, 1.4, 1.6, 1.5, 1.7, 1.2, 1.6, 1.7, 1.8,
    1.4, 1.5, 1.6, 1.3, 1.1, 1.4, 1.5, 1.8, 1.7, 1.6,
    1.4, 1.2, 1.3, 1.5, 1.7, 1.6, 1.8, 1.5, 1.9, 1.8
])

# 1. Hitung total sampah yang dikumpulkan di masing-masing wilayah selama sebulan.
total_wilayah1 = np.sum(sampah_wilayah1)
total_wilayah2 = np.sum(sampah_wilayah2)

print(f"Total sampah di Wilayah 1 selama sebulan: {total_wilayah1} ton")
print(f"Total sampah di Wilayah 2 selama sebulan: {total_wilayah2} ton")
```

```python
# 2. Tentukan hari di mana wilayah pertama mengumpulkan lebih banyak sampah daripada wilayah kedua dan sebaliknya.
hari_wilayah1_lebih_banyak = np.where(sampah_wilayah1 > sampah_wilayah2)[0]
hari_wilayah2_lebih_banyak = np.where(sampah_wilayah2 > sampah_wilayah1)[0]

print(f"wilayah 1 mengumpulkan sampah lebih banyak di hari : {hari_wilayah1_lebih_banyak + 1}")
print(f"wilayah 2 mengumpulkan sampah lebih banyak di hari : {hari_wilayah2_lebih_banyak + 1}")
```

```python
# 3. Hitung total sampah yang dikumpulkan oleh kedua wilayah setiap harinya.
total_per_hari = sampah_wilayah1 + sampah_wilayah2
print(f"Total sampah yang dikumpulkan oleh kedua wilayah setiap harinya: {total_per_hari}")
```

```python
import pandas as pd

# Data sampah dari tabel dalam bentuk dictionary
data_sampah = {
    'Kabupaten/Kota': [
        "Kota Bandung", "Kota Bekasi", "Kota Bogor", "Kota Depok", "Kota Cimahi", 
        "Kabupaten Bandung", "Kabupaten Bekasi", "Kabupaten Bogor", "Kabupaten Cirebon", "Kabupaten Garut",
        "Kabupaten Indramayu", "Kabupaten Karawang", "Kabupaten Kuningan", "Kabupaten Majalengka",
        "Kabupaten Pangandaran", "Kabupaten Purwakarta", "Kabupaten Subang", "Kabupaten Sukabumi",
        "Kabupaten Sumedang", "Kabupaten Tasikmalaya"
    ],
    'Produksi Sampah 2020 (ton)': [
        280000, 500000, 240000, 210000, 120000, 350000, 410000, 370000, 220000, 260000, 
        300000, 450000, 200000, 190000, 140000, 230000, 250000, 310000, 180000, 270000
    ],
    'Produksi Sampah 2021 (ton)': [
        290000, 510000, 250000, 220000, 125000, 355000, 415000, 375000, 225000, 265000, 
        305000, 455000, 205000, 195000, 145000, 235000, 255000, 315000, 185000, 275000
    ],
    'Produksi Sampah 2022 (ton)': [
        300000, 520000, 260000, 230000, 130000, 360000, 420000, 380000, 230000, 270000, 
        310000, 460000, 210000, 200000, 150000, 240000, 260000, 320000, 190000, 280000
    ]
}

# Membuat DataFrame
df_sampah = pd.DataFrame(data_sampah)

# 1. Menghitung total produksi sampah per tahun
total_2020 = df_sampah['Produksi Sampah 2020 (ton)'].sum()
total_2021 = df_sampah['Produksi Sampah 2021 (ton)'].sum()
total_2022 = df_sampah['Produksi Sampah 2022 (ton)'].sum()

print(f"Total Produksi Sampah 2020: {total_2020} ton")
print(f"Total Produksi Sampah 2021: {total_2021} ton")
print(f"Total Produksi Sampah 2022: {total_2022} ton")

# 2. Menghitung total produksi sampah per Kabupaten/Kota per tahun menggunakan iterrows
df_sampah['Total Produksi Sampah Per Tahun'] = 0

for index, row in df_sampah.iterrows():
    total_per_kota = row['Produksi Sampah 2020 (ton)'] + row['Produksi Sampah 2021 (ton)'] + row['Produksi Sampah 2022 (ton)']
    df_sampah.at[index, 'Total Produksi Sampah Per Tahun'] = total_per_kota

# Menampilkan DataFrame yang telah dihitung
print("\nData dengan Total Produksi Sampah Per Tahun:")
print(df_sampah)

# Mengekspor hasil akhir ke file CSV dan Excel
df_sampah.to_csv("produksi_sampah_jabar_total.csv", index=False)
df_sampah.to_excel("produksi_sampah_jabar_total.xlsx", index=False)

df_sampah
```
