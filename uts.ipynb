```python
import matplotlib.pyplot as plt

def draw_circle(x_center, y_center, radius):
    # Inisialisasi variabel
    x = 0
    y = radius
    d = 1 - radius  # Keputusan awal

    points = []

    # Fungsi untuk mencerminkan titik-titik dari kuadran pertama ke seluruh lingkaran
    def plot_circle_points(x_center, y_center, x, y):
        points.extend([
            (x_center + x, y_center + y),
            (x_center - x, y_center + y),
            (x_center + x, y_center - y),
            (x_center - x, y_center - y),
            (x_center + y, y_center + x),
            (x_center - y, y_center + x),
            (x_center + y, y_center - x),
            (x_center - y, y_center - x)
        ])

    # Menghitung titik-titik lingkaran
    plot_circle_points(x_center, y_center, x, y)
    while x < y:
        if d < 0:
            d += 2 * x + 3
        else:
            d += 2 * (x - y) + 5
            y -= 1
        x += 1
        plot_circle_points(x_center, y_center, x, y)

    return points

# Pusat lingkaran dan jari-jari
x_center, y_center = 0, 0
radius = 9

# Mendapatkan titik-titik lingkaran dengan algoritma
circle_points = draw_circle(x_center, y_center, radius)

# Menampilkan hasil dengan Matplotlib dalam warna biru
x_coords, y_coords = zip(*circle_points)  # Pisahkan koordinat x dan y
plt.plot(x_coords, y_coords, marker="o", color="blue", linestyle="None")
plt.gca().set_aspect('equal', adjustable='box')  # Menjaga aspek rasio agar lingkaran terlihat sempurna
plt.grid(True)
plt.title("Lingkaran dengan Jari-jari 9 Berwarna Biru")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import pygame
import random

# Inisialisasi Pygame
pygame.init()

# Dimensi layar
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Akuarium Virtual 2D")

# Warna
BLUE = (0, 105, 148)
WHITE = (255, 255, 255)

# Ikan (lingkaran sederhana)
fish_positions = [
    [random.randint(50, WIDTH-50), random.randint(50, HEIGHT-50)] for _ in range(5)
]
fish_speed = [random.randint(1, 3) for _ in range(5)]

# Fungsi menggambar ikan
def draw_fish(x, y):
    pygame.draw.ellipse(screen, WHITE, (x, y, 40, 20))  # Bentuk ikan sederhana
    pygame.draw.polygon(screen, WHITE, [(x+40, y+10), (x+50, y+5), (x+50, y+15)])  # Sirip ikan

# Loop utama
running = True
clock = pygame.time.Clock()

while running:
    screen.fill(BLUE)  # Latar belakang laut

    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Gerakan ikan
    for i, pos in enumerate(fish_positions):
        pos[0] += fish_speed[i]  # Gerak horizontal
        if pos[0] > WIDTH:  # Jika ikan keluar layar
            pos[0] = -50  # Kembali dari kiri layar
        draw_fish(pos[0], pos[1])  # Gambar ikan

    pygame.display.flip()
    clock.tick(30)

pygame.quit()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk membuat bintang
def create_star(radius1, radius2, num_points):
    angles = np.linspace(0, 2 * np.pi, num_points * 2, endpoint=False)
    radii = np.array([radius1 if i % 2 == 0 else radius2 for i in range(len(angles))])
    x = radii * np.cos(angles)
    y = radii * np.sin(angles)
    return np.column_stack((x, y))

# Bintang awal
original_star = create_star(radius1=1, radius2=0.5, num_points=5)

# Translasi
translation_vector = np.array([2, 3])
translated_star = original_star + translation_vector

# Refleksi terhadap y = x
reflected_star = np.column_stack((original_star[:, 1], original_star[:, 0]))

# Rotasi 45 derajat (pi/4 radian)
theta = np.pi / 4
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta), np.cos(theta)]
])
rotated_star = original_star @ rotation_matrix.T

# Plot hasil
plt.figure(figsize=(10, 10))

# Plot bintang asli
plt.plot(original_star[:, 0], original_star[:, 1], label='Original Star', marker='o')

# Plot translasi
plt.plot(translated_star[:, 0], translated_star[:, 1], label='Translated Star', marker='o')

# Plot refleksi
plt.plot(reflected_star[:, 0], reflected_star[:, 1], label='Reflected Star (y=x)', marker='o')

# Plot rotasi
plt.plot(rotated_star[:, 0], rotated_star[:, 1], label='Rotated Star (45)', marker='o')

# Pengaturan plot
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.title('2D Transformations: Star Shape')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axis('equal')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Fungsi untuk menggambar segitiga
def draw_triangle(ax, vertices, color):
    triangle = Poly3DCollection([vertices], color=color, alpha=0.5)
    ax.add_collection3d(triangle)

# Matriks refleksi untuk berbagai bidang
def reflection_matrix_xy():
    return np.array([
        [1, 0, 0],
        [0, 1, 0],
        [0, 0, -1]
    ])

def reflection_matrix_yz():
    return np.array([
        [-1, 0, 0],
        [0, 1, 0],
        [0, 0, 1]
    ])

def reflection_matrix_xz():
    return np.array([
        [1, 0, 0],
        [0, -1, 0],
        [0, 0, 1]
    ])

# Fungsi utama untuk menggambar objek 3D dengan refleksi
def plot_3d():
    # Titik-titik untuk segitiga
    vertices = np.array([
        [0, 0.5, 0],  # puncak segitiga
        [-0.5, -0.5, 0.5],  # kiri bawah
        [0.5, -0.5, 0.5],  # kanan bawah
        
        [0, 0.5, 0],  # puncak segitiga belakang
        [0.5, -0.5, -0.5],  # kanan bawah belakang
        [-0.5, -0.5, -0.5]  # kiri bawah belakang
        
        # segitiga lainnya bisa ditambahkan jika diperlukan
    ])

    # Membuat figure dan axis
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Gambar segitiga asli
    draw_triangle(ax, vertices[:3], 'red')
    draw_triangle(ax, vertices[3:6], 'green')
    ax.view_init(elev=30, azim=45)

    # Gambar refleksi segitiga
    vertices_reflected = np.dot(vertices, reflection_matrix_xz())
    draw_triangle(ax, vertices_reflected[:3], 'blue')
    draw_triangle(ax, vertices_reflected[3:6], 'yellow')

    # Menambahkan grid dan tampilan
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])
    plt.show()

# Menjalankan fungsi untuk menggambar objek 3D
if __name__ == "__main__":
    plot_3d()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Fungsi untuk menggambar balok
def draw_cube(ax, vertices, color):
    # Mendefinisikan sisi-sisi balok menggunakan vertices
    faces = [
        [vertices[0], vertices[1], vertices[2], vertices[3]],  # sisi depan
        [vertices[4], vertices[5], vertices[6], vertices[7]],  # sisi belakang
        [vertices[0], vertices[1], vertices[5], vertices[4]],  # sisi kiri
        [vertices[2], vertices[3], vertices[7], vertices[6]],  # sisi kanan
        [vertices[0], vertices[3], vertices[7], vertices[4]],  # sisi atas
        [vertices[1], vertices[2], vertices[6], vertices[5]],  # sisi bawah
    ]
    cube = Poly3DCollection(faces, color=color, alpha=0.5)
    ax.add_collection3d(cube)

# Matriks refleksi untuk berbagai bidang
def reflection_matrix_xy():
    return np.array([
        [1, 0, 0],
        [0, 1, 0],
        [0, 0, -1]
    ])

def reflection_matrix_yz():
    return np.array([
        [-1, 0, 0],
        [0, 1, 0],
        [0, 0, 1]
    ])

def reflection_matrix_xz():
    return np.array([
        [1, 0, 0],
        [0, -1, 0],
        [0, 0, 1]
    ])

# Fungsi utama untuk menggambar objek 3D dengan refleksi
def plot_3d():
    # Titik-titik untuk balok
    vertices = np.array([
        [-0.5, -0.5, -0.5],  # kiri bawah depan
        [0.5, -0.5, -0.5],  # kanan bawah depan
        [0.5, 0.5, -0.5],   # kanan atas depan
        [-0.5, 0.5, -0.5],  # kiri atas depan
        
        [-0.5, -0.5, 0.5],  # kiri bawah belakang
        [0.5, -0.5, 0.5],   # kanan bawah belakang
        [0.5, 0.5, 0.5],    # kanan atas belakang
        [-0.5, 0.5, 0.5],   # kiri atas belakang
    ])

    # Membuat figure dan axis
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Gambar balok asli
    draw_cube(ax, vertices, 'red')

    # Gambar refleksi balok
    vertices_reflected = np.dot(vertices, reflection_matrix_xz())  # Menggunakan refleksi di bidang XZ
    draw_cube(ax, vertices_reflected, 'blue')

    # Menambahkan grid dan tampilan
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])
    ax.view_init(elev=30, azim=45)  # Menyesuaikan tampilan
    plt.show()

# Menjalankan fungsi untuk menggambar objek 3D
if __name__ == "__main__":
    plot_3d()
```

```python
import pygame
import math

# Inisialisasi pygame
pygame.init()

# Dimensi layar
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Akuarium Virtual")

# Warna
BLUE = (0, 105, 148)
WHITE = (255, 255, 255)
YELLOW = (255, 223, 0)

# Fungsi untuk menggambar ikan sederhana
def draw_fish(surface, x, y, scale, angle):
    # Transformasi translasi, skala, dan rotasi
    fish_body = [
        (0, 0), (50, 20), (80, 0), (50, -20), (0, 0)  # Bentuk tubuh ikan
    ]
    tail = [
        (-20, 10), (-40, 0), (-20, -10)  # Ekor ikan
    ]

    # Skala dan rotasi
    rotated_body = [
        (
            scale * (point[0] * math.cos(angle) - point[1] * math.sin(angle)),
            scale * (point[0] * math.sin(angle) + point[1] * math.cos(angle))
        ) for point in fish_body
    ]
    rotated_tail = [
        (
            scale * (point[0] * math.cos(angle) - point[1] * math.sin(angle)),
            scale * (point[0] * math.sin(angle) + point[1] * math.cos(angle))
        ) for point in tail
    ]

    # Translasi
    translated_body = [(x + point[0], y + point[1]) for point in rotated_body]
    translated_tail = [(x + point[0], y + point[1]) for point in rotated_tail]

    pygame.draw.polygon(surface, YELLOW, translated_body)
    pygame.draw.polygon(surface, WHITE, translated_tail)

# Fungsi utama
def main():
    clock = pygame.time.Clock()
    running = True

    # Posisi awal ikan
    fish_x, fish_y = 400, 300
    angle = 0
    scale = 1

    while running:
        screen.fill(BLUE)  # Latar belakang akuarium

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        # Animasi translasi, rotasi, dan skala
        fish_x += 2  # Translasi horizontal
        if fish_x > WIDTH + 50:
            fish_x = -50

        angle += 0.05  # Rotasi
        if angle > 2 * math.pi:
            angle -= 2 * math.pi

        scale = 1 + 0.5 * math.sin(pygame.time.get_ticks() / 1000)  # Skala animasi

        # Gambar ikan
        draw_fish(screen, fish_x, fish_y, scale, angle)

        # Update layar
        pygame.display.flip()
        clock.tick(60)

    pygame.quit()

if __name__ == "__main__":
    main()
```

```python
import pygame
import math
import random

# Inisialisasi pygame
pygame.init()

# Dimensi layar
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Akuarium Virtual")

# Warna
BLUE = (0, 105, 148)
WHITE = (255, 255, 255)
YELLOW = (255, 223, 0)
RED = (255, 69, 0)
GREEN = (50, 205, 50)
ORANGE = (255, 165, 0)

# Fungsi untuk menggambar ikan realistis
def draw_realistic_fish(surface, x, y, scale, angle, body_color, eye_color):
    # Tubuh ikan (oval)
    body_width = 80 * scale
    body_height = 40 * scale
    body = pygame.Surface((body_width, body_height), pygame.SRCALPHA)
    pygame.draw.ellipse(body, body_color, (0, 0, body_width, body_height))

    # Sirip belakang (segitiga)
    tail = pygame.Surface((body_width, body_height), pygame.SRCALPHA)
    pygame.draw.polygon(tail, body_color, [
        (0, body_height // 2), (-20 * scale, 0), (-20 * scale, body_height)
    ])

    # Mata ikan (lingkaran kecil)
    eye_x = int(body_width * 0.75)
    eye_y = int(body_height * 0.35)
    eye_radius = int(5 * scale)
    pygame.draw.circle(body, eye_color, (eye_x, eye_y), eye_radius)
    pygame.draw.circle(body, WHITE, (eye_x, eye_y), eye_radius // 2)

    # Rotasi tubuh dan sirip
    rotated_body = pygame.transform.rotate(body, math.degrees(-angle))
    rotated_tail = pygame.transform.rotate(tail, math.degrees(-angle))

    # Posisi tubuh dan sirip
    body_rect = rotated_body.get_rect(center=(x, y))
    tail_rect = rotated_tail.get_rect(center=(x - 30 * scale, y))

    # Gambar ke layar
    surface.blit(rotated_tail, tail_rect.topleft)
    surface.blit(rotated_body, body_rect.topleft)

# Fungsi untuk menggambar gelembung udara
def draw_bubble(surface, x, y, radius):
    pygame.draw.circle(surface, WHITE, (x, y), radius, 2)

# Fungsi utama
def main():
    clock = pygame.time.Clock()
    running = True

    # Posisi dan atribut ikan
    fish_positions = [
        {
            "x": random.randint(0, WIDTH), 
            "y": random.randint(100, HEIGHT - 100), 
            "scale": random.uniform(0.5, 1.5), 
            "speed": random.uniform(1, 3), 
            "angle": 0, 
            "body_color": random.choice([YELLOW, ORANGE, GREEN]), 
            "eye_color": random.choice([RED, WHITE])
        } 
        for _ in range(5)
    ]

    # Gelembung
    bubbles = [{"x": random.randint(0, WIDTH), "y": random.randint(HEIGHT, HEIGHT + 200), "radius": random.randint(5, 15)} for _ in range(20)]

    while running:
        screen.fill(BLUE)  # Latar belakang akuarium

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        # Animasi ikan
        for fish in fish_positions:
            fish["x"] += fish["speed"]  # Translasi horizontal
            if fish["x"] > WIDTH + 50:
                fish["x"] = -50
                fish["y"] = random.randint(100, HEIGHT - 100)  # Reset posisi vertikal

            fish["angle"] += 0.05  # Rotasi
            if fish["angle"] > 2 * math.pi:
                fish["angle"] -= 2 * math.pi

            # Gambar ikan
            draw_realistic_fish(screen, fish["x"], fish["y"], fish["scale"], fish["angle"], fish["body_color"], fish["eye_color"])

        # Animasi gelembung
        for bubble in bubbles:
            bubble["y"] -= 2  # Gelembung naik ke atas
            if bubble["y"] < -20:
                bubble["y"] = random.randint(HEIGHT, HEIGHT + 200)
                bubble["x"] = random.randint(0, WIDTH)  # Reset posisi horizontal

            draw_bubble(screen, bubble["x"], bubble["y"], bubble["radius"])

        # Update layar
        pygame.display.flip()
        clock.tick(60)

    pygame.quit()

if __name__ == "__main__":
    main()
```
